```python
import requests

API_URL = "https://api-inference.huggingface.co/models/finiteautomata/bertweet-base-sentiment-analysis"
headers = {"Authorization": "Bearer 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"}

def sentiment_analysis_api(payload):
	response = requests.post(API_URL, headers=headers, json=payload)
	return response.json()

output = sentiment_analysis_api({
	"inputs": news_data
})

output
```

```python
list_turple = []
for i in output:
    d = {}
    for j in i:
        d[j['label']] = j['score']
        
    d["sentiment_analysis_result"] = list(d.keys())[0]
    list_turple.append(d)
    d = {}
    
list_turple
```

```python
import pandas as pd
final_dataframe = pd.DataFrame()
final_dataframe['Textile_news'] = news_data
final_dataframe['POSITIVE'] = [ round(i['POS'] * 100,2) for i in list_turple]
final_dataframe['NEUTRAL'] = [ round(i['NEU'] * 100,2) for i in list_turple]
final_dataframe['NEGATIVE'] = [ round(i['NEG'] * 100,2) for i in list_turple]
map_sentiment = {'POS':'POSITIVE','NEU':'NEUTRAL','NEG':'NEGATIVE'}
final_dataframe['sentiment_analysis_result'] = [ map_sentiment[i['sentiment_analysis_result']] for i in list_turple]
```
